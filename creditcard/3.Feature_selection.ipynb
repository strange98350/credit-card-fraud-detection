```python
import pandas as pd
```

```python
df=pd.read_csv("undersampled_data.csv")
df
```

```python
df['Amount'] = df['Amount'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
col = ['UseChip', 'MerchantName', 'MerchantCity', 'MerchantCountry','target']

for i in col:
    df[i] = le.fit_transform(df[i]).astype(int)
```

```python
independent=df[['User', 'Card', 'Year', 'Month', 'Day','UseChip','Amount',
       'MerchantName', 'MerchantCity', 'MerchantCountry', 'Zip', 'MCC']]

dependent=df[['target']]
independent
```

# Check the correlation for balanced dataset
# correlation -relationship between 2 variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(25,25))
sns.heatmap(df.corr(),annot=True,ax=ax)
plt.show()
```

# Feature Selection

```python
cor=df.corr()
cor_target=abs(cor["target"])
relevant_features=cor_target[cor_target>0.06]
print(relevant_features)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import numpy as np

# Create the ExtraTreesClassifier model
model = ExtraTreesClassifier()

# Fit the model on your data
model.fit(independent, dependent)

# Get feature importances
importances = model.feature_importances_

# Sort the feature importances in descending order
sorted_indices = np.argsort(importances)[::-1]

# Plot feature importances
plt.figure(figsize=(12, 8))
plt.title('Feature Importances')
plt.bar(range(independent.shape[1]), importances[sorted_indices], align='center')
plt.xticks(range(independent.shape[1]), independent.columns[sorted_indices], rotation=90)
plt.tight_layout()
plt.show()

# Print feature importances
for i, index in enumerate(sorted_indices):
    print(f"{i + 1}. Feature: {independent.columns[index]} - Importance: {importances[index]}")
```
